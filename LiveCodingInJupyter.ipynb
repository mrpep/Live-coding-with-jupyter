```python
from audio_engine import AudioEngine
from signals import Sequencer, Oscillator
import threading

audio_engine = AudioEngine()

audio_thread = threading.Thread(target = audio_engine.run, daemon = True)
audio_thread.start()
audio_engine.play()
```

```python
fm2 = Oscillator(freq=20,amp=0.1,phase=0,offset=440)
fm1 = Oscillator(freq=1,amp=1,phase=0,offset=fm2)
osc1 = Oscillator(freq=fm1,amp=1,phase=0)
osc2 = Oscillator(freq=osc1,amp=0.5,phase=fm2)
audio_engine.add_sound(osc2)
```
